```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##sentences
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'i am good girl',
    'i am good developer',
    'understand the meaning of words',
    'your videos are good',
]
```

```python
sent
```

```python
## define the vocab size
voc_size=10000
```

```python
##one hot reps
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
#word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
#we make all the sentences of equal size
sent_length=8#max sent length
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)#pad seq basically jitte word kam rehte utte ya toh aage ya peeche 0 lga dete
```

```python
embedded_docs
```

```python
#feature reps
dim=10#feature reps ki dimension
```

```python
model=Sequential()
#model wrt embedding layer
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
